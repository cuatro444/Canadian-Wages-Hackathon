```python
# Plotting and Data Handling 
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.tree import plot_tree


# Modeling
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

## Import the Data

```python
data = pd.read_excel('AvgWages.xlsx')
data.head(10)
```

## Prepare the Data

#### Employed and Unemployed

```python
dataHourlyEmp = data[data['Wages'] == 'Average hourly wage rate']
dataHourlyEmp = dataHourlyEmp[dataHourlyEmp['Value'] != 0]
dataHourlyUnEmp = dataHourlyEmp[dataHourlyEmp['Value'] == 0]
dataHourlyEmp.head(10)
dataHourlyEmp['Value'].mean()
```

#### Randomnize the Data

```python
# Randomnize the data First
dataHourlyEmp = dataHourlyEmp.sample(frac=1).reset_index(drop=True)
dataHourlyEmp['Value'].mean()
```

#### Label Enconding

#### Split into X and y

```python
# Split into X and y
y = dataHourlyEmp['Value']
X = dataHourlyEmp.drop(columns = ['Value', 'Wages']).apply(LabelEncoder().fit_transform)
X.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=.2, random_state = 1)
```

## Modeling

#### Quick Simple DT

```python
# Define
model_DT = DecisionTreeRegressor()

# Fit
model_DT.fit(X_train, y_train)

# Predict
y_test_pred = model_DT.predict(X_test)

# Evaluate 
mae = mean_absolute_error(y_test, y_test_pred)

print('MAE (after train-test-split):')
print(mae)
```

#### Optimize Criterion and Depth 

```python
# Set the range for depths
depths = list(range(2, 50, 2))

for depth in depths:
    print(depth)
    
    # Define    
    model_DT = DecisionTreeRegressor(max_depth = depth)
        
    # Fit
    model_DT.fit(X_train, y_train)

    # Predict
    y_test_pred = model_DT.predict(X_test)

    # Evaluate 
    mae = mean_absolute_error(y_test, y_test_pred)
    
    print(mae)
    
    
```

#### Plot the Decision Tree

```python
plt.figure(figsize=(20,10), dpi=100)
plot_tree(model_DT, max_depth=3, feature_names=X.columns, fontsize=6, filled=True)
plt.show()
```
