### Comparison Between Sexes

### Comparison Between Sexes Value Range(100, 1500) for Vizualization Weekly Wage Rate Mean

```python
male_value = data[data['Sex'] == '  Male']
female_value = data[data['Sex'] == '  Female']

male_weekly = male_value[male_value['Wages'] == 'Average weekly wage rate']
female_weekly = female_value[female_value['Wages'] == 'Average weekly wage rate']

male_value = male_value[male_value['Wages'] == 'Average weekly wage rate']['Value']
female_value = female_value[female_value['Wages'] == 'Average weekly wage rate']['Value']

plt.figure(figsize = (20, 10))
plt.hist(male_value, bins = 50, range = (50, 1500), alpha=0.6, label = 'Male')
plt.hist(female_value, bins = 50, range = (50, 1500),  alpha=0.6, color = 'pink', label = 'Female')
plt.legend(loc='upper right')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range 50 - 1500')
plt.show()

plt.figure(figsize = (20, 10))
plt.hist(male_value, bins = 50, range = (0, 50), alpha=0.6, label = 'Male')
plt.hist(female_value, bins = 50, range = (0, 50),  alpha=0.6, color = 'pink', label = 'Female')
plt.legend(loc='upper right')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range 0 - 50')
plt.show()

plt.figure(figsize = (20, 10))
plt.hist(male_value, bins = 50, range = (1500, 3000), alpha=0.6, label = 'Male')
plt.hist(female_value, bins = 50, range = (1500,3000),  alpha=0.6, color = 'pink', label = 'Female')
plt.legend(loc='upper right')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Range 1500 -3000')
plt.show()
```

We notice a significant difference in the distribution for male vs female workers. It appears more women are lumped into the lower salaries and then the distribution decreases. in the case of male workers, we notice that there are two peaks as opposed to one decrease in the distribution. We excluded the outliars for the purpose of vizualization.

```python
### Just go over all the columns lol 
columns = data.columns
```

```python
for column in columns: 
    ColMales = dict(data[data['Sex'] == '  Male'][column].value_counts())
    ColFemales = dict(data[data['Sex'] == '  Female'][column].value_counts())
    
    # Get values from dictionary
    res_Males= []
    res_Females = []
    for key in ColMales.keys():
        res_Males.append(ColMales[key])
    for key in ColFemales.keys():
        res_Females.append(ColFemales[key])
        
    print(column, '\n')
    plt.figure(figsize = (20, 10))
    plt.bar(list(ColMales.keys()), res_Males,  alpha=0.6, label = 'Male')
    plt.bar(list(ColFemales.keys()), res_Females, alpha=0.6, color = 'pink', label = 'Female')
    plt.legend(loc='upper right')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.show()
```
